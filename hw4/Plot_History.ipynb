```python
import torch
import matplotlib.pyplot as plta
```

```python
device = torch.device('cuda')
```

```python
continue_from = "checkpoint/Agent480"
if str(device) == 'cpu':
    saved_model = torch.load(continue_from, map_location=lambda storage, location: storage)
else:
    saved_model = torch.load(continue_from)
```

```python
reward_history = saved_model['reward_history']
loss_history = saved_model['loss_history']
```

```python
plta.figure(figsize=(15,4))
plta.plot(reward_history, label='reward')
plta.xlabel('episode')
plta.ylabel('reward')
plta.title('history')

plta.legend(loc='upper right')
plta.show()
```
